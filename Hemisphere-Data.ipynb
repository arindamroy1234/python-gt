```python
%matplotlib inline

import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm

whole_data = pd.read_csv("C:\\Users\\arind\\python project\\data-visualization\\ExcelFormattedGISTEMPData2CSV.csv")
whole_data.head()
```

```python
df = whole_data.copy()
df
```

```python
import seaborn as sns
sns.set()
sns.set_context('notebook', font_scale=1.2)
sns.set_style('ticks')
sns.set(rc={'figure.figsize':(15.7,12.27)})
```

```python
# construct cmap
from matplotlib.colors import ListedColormap
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
my_cmap = ListedColormap(sns.color_palette(flatui).as_hex())
df.plot(kind='bar', x='Year', y='Glob', colormap=my_cmap)
plt.ylabel("Temperature[Celcius]")
plt.title("Global Temperature Increment over years")
```

```python
#df.set_index(["Year"],inplace=True)
df.loc[1880].plot(kind='bar', figsize=(6,4), rot=30, color=cm.tab10.colors[0:7])
plt.ylabel("Temperature[Celcius]")
plt.title("Segment Data around the Globe - Yr 1880")
```

```python
df.loc[2014].plot(kind='bar', figsize=(6,4), rot=30, color=cm.tab10.colors[0:7])
plt.ylabel("Temperature[Celcius]")
plt.title("Segment Data around the Globe - Yr 2014")
```

```python
tdf = df.T
```

```python
tdf
```

```python
month = list()
sns.set_context('notebook', font_scale=1.1)
sns.set_style('ticks')
sns.set(rc={'figure.figsize':(15.7,12.27)})
plt.xlabel("Year")
plt.ylabel("Temperature[Celcius]")
plt.title("Yearly temperature increase in Celcius")
for col in df.columns:
    if not col == 'Year':
        sns.lineplot(data=df, x='Year', y=col, estimator=None)
        month.append(col)
        plt.legend(month)
```

```python
#tdf.loc['Glob'].plot(kind='bar', figsize=(12,10), color=cm.tab10.colors[0:5])
sns.set_context('notebook', font_scale=1.2)
sns.set_style('ticks')
sns.set(rc={'figure.figsize':(15.7,12.27)})
tdf.plot(kind='line', color=cm.tab10.colors[0:5])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.ylabel("Temperature[Celcius]")
plt.title("Segment Data Yearwise")
```
