```python
#sns.lineplot(data=df1, x='Year', y='Jan' )
#sns.lineplot(data=df1, x='Year', y='Feb' )
month = list()
sns.set_context('notebook', font_scale=1.1)
sns.set_style('ticks')
sns.set(rc={'figure.figsize':(15.7,12.27)})
plt.xlabel("Year")
plt.ylabel("Temperature[Celcius]")
plt.title("Yearly temperature increase in Celcius monthwise data")
for col in df1.columns:
    if not col == 'Year':
        sns.lineplot(data=df1, x='Year', y=col)
        month.append(col)
        plt.legend(month)
#sns.lineplot(data=df1 )
```
